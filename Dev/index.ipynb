Fonctions - Personal set of function to use in the below suit of program

```python
# Read float from standard input

def read_float(textToPrint):
    fNumber = float('inf')
    while fNumber == float('inf'):
        x = input(textToPrint)
        try:
            fNumber = float(x)
        except ValueError:
            print("Your enter is not a float !")
    return fNumber

# Read a float number as denominator from a standard input
def read_float_denominator(textToPrint):
    fdenominator = 0
    while fdenominator == 0:
        x = input(textToPrint)
        try:
            fdenominator = float(x)
        except ValueError:
            print("Your enter is not a float !")
    return fdenominator


# Read integer from standard input

def read_int(textToPrint):
    fNumber = float('inf')
    while fNumber == float('inf'):
        x = input(textToPrint)
        try:
            fNumber = int(x)
        except ValueError:
            print("Your enter is not a integer !")
    return fNumber
```

Program 02 - This program implement the division of two numbers passed in standard input

```python
# Read numerator
numerator = read_float("Enter the numerator, it should be a float: ")

# Read denominator
denominator = read_float_denominator("Enter the denominator, it should be a float different of 0: ")

# Division operation
ratio = numerator/denominator
print("{0} / {1} = {2}".format(numerator, denominator, ratio))


```
