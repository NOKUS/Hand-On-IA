### Program 02
#### This program implement the division of two numbers passed in standard input

```python
# Read numerator
numerator = float('inf')
while numerator == float('inf'):
    x = input("Enter the numerator, it should be a float: ")
    try:
        numerator = float(x)
    except ValueError:
        print("Your entre is not a float")

# Read denominator
denominator = 0

while denominator == 0:
    x = input("Enter the denominator, it should be a float different of 0: ")
    try:
        denominator = float(x)
    except ValueError:
        print("Your entre is not a float or is 0")

# Division operation
ratio = numerator/denominator
print("{0} / {1} = {2}".format(numerator, denominator, ratio))


```

#### Program 03
##### compute Air of a triangle

```python
# Read base
base = float('inf')
while base == float('inf'):
    x = input("Enter the base of your traingle (float in cm): ")
    try:
        base = float(x)
    except ValueError:
        print("Your entre is not a float !")

# Read height
height = float('inf')
while height == float('inf'):
    x = input("Enter the height of your traingle (float in cm): ")
    try:
        height = float(x)
    except ValueError:
        print("Your entre is not a float !")

print("The Air of your Triangle is : {0} cm^2".format(base*height/2.0))
```
